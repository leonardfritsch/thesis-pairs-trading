```python
import pandas as pd
import yfinance as yf
import numpy as np
import matplotlib.pyplot as plt
import datetime
```

```python
vix = yf.download('^VIX')['Adj Close']
```

```python
plt.plot(vix)
```

```python
vix_2000 = vix.loc['2000-01-01':]
plt.plot(vix_2000)
```

```python
vix_sorted = vix_2000.sort_values(ascending=False)
print(vix_sorted)
```

```python
vix_sorted = vix_sorted[:int((len(vix_sorted)*0.03))]
print(vix_sorted)
```

```python
index = vix_sorted.index
```

```python
print(index)
```

```python
years = index.year
print(years)
```

```python
years = years.unique().values.tolist()
print(years)
```

```python
vix_sorted['2020'].index
```

```python
for i in years:
    print(vix_sorted[str(i)].index[0], vix_sorted[str(i)].max())
```

```python
max_vals = []
date = []
for i in years:
    max_vals.append(vix_sorted[str(i)].max())
    date.append(vix_sorted[str(i)].index[0])
    
s = pd.Series(max_vals, index= date)
s
```
